## Let us now build some models

### First, let's build a simple KNN model:

### First, we'd want to split and more importantly SCALE the data, so that no variable has any extra weight than any other variable based on their own scales

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(Xdata,ydatamulti,random_state=0)


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Xtrainscaled = scaler.fit_transform(Xtrain)
Xtestscaled = scaler.transform(Xtest)
```

```python
from sklearn.neighbors import KNeighborsClassifier

for k in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(Xtrainscaled,ytrain)
    print(k)
    print("Training Accuracy:")
    print(knn.score(Xtrainscaled,ytrain))
    print("Test Accuracy:")
    print(knn.score(Xtestscaled,ytest))
```

### It's clear that testing accuracy performs worse on any k larger than 9. Between k=1,9 we have a few with a testing accuracy of 66%. One could argue the simplest k would be best, but with a k=3, we have a relatively high test accuracy, and a training accuracy that is similar, meaning we are most likely not under or overfitting. This model would best generalize into new datasets. Still, the accuracy is quite low overall, so this model may perform better within an ensemble model.

```python
knnFinal = KNeighborsClassifier(n_neighbors=3)
knnFinal.fit(Xtrainscaled,ytrain)
print('k:',3)
print("Training Accuracy:")
print(knnFinal.score(Xtrainscaled,ytrain))
print("Test Accuracy:")
print(knnFinal.score(Xtestscaled,ytest))
```

### Next, let's try a Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

Xtrain,Xtest,ytrain,ytest = train_test_split(Xdata,ydatamulti,random_state=0)


for d in range (1,20):
    tree = DecisionTreeClassifier(max_depth = d).fit(Xtrain,ytrain)
    print(d)
    print(tree.score(Xtrain,ytrain))
    print(tree.score(Xtest,ytest))
```

### When we first ran the decision tree classifier without giving a max decision tree depth limit, the test accuracy was in the low 60s. Looking at the above results, we see that any depth greater than 8 results in overfitting of the model. Looking at the test accuracy, a maximum depth of 7 or 8 would best generalize to new datasets.

```python
treeFinal = DecisionTreeClassifier(max_depth = 7)
treeFinal.fit(Xtrain,ytrain)
print('Max Depth:',7)
print('Training Score:',treeFinal.score(Xtrain,ytrain))
print('Testing Score:',treeFinal.score(Xtest,ytest))
```

### Next, let's examine a linear SVM classification model

```python
from sklearn.svm import LinearSVC

Xtrain,Xtest,ytrain,ytest = train_test_split(Xdata,ydatamulti,random_state=0)

scaler = MinMaxScaler()
Xtrainscaled = scaler.fit_transform(Xtrain)
Xtestscaled = scaler.transform(Xtest)

carray = np.linspace(0.1,1,10)
for c in carray:
    LSVM = LinearSVC(C=c,random_state=67).fit(Xtrainscaled,ytrain)
    print("C",c)
    print("training score",LSVM.score(Xtrainscaled,ytrain))
    print("testing score",LSVM.score(Xtestscaled,ytest))
for c in range(2,10):
    LSVM = LinearSVC(C=c,random_state=67).fit(Xtrainscaled,ytrain)
    print("C",c)
    print("training score",LSVM.score(Xtrainscaled,ytrain))
    print("testing score",LSVM.score(Xtestscaled,ytest))
```

### The best training and testing scores result from a C around 0.7

```python
LSVMFinal = LinearSVC(C=0.7,random_state=67)
LSVMFinal.fit(Xtrainscaled,ytrain)
print("C",0.7)
print("training score",LSVM.score(Xtrainscaled,ytrain))
print("testing score",LSVM.score(Xtestscaled,ytest))
```

### Next, let's try out a kernelized SVM

```python
from sklearn.svm import SVC

Xtrain,Xtest,ytrain,ytest = train_test_split(Xdata,ydatamulti,random_state=0)

scaler = MinMaxScaler()
Xtrainscaled = scaler.fit_transform(Xtrain)
Xtestscaled = scaler.transform(Xtest)

garray = (0.1,1,10,50,100)
carray = (0.1,1,10,50,100)
for c in carray:
    for g in garray:
        KSVM = SVC(kernel = 'rbf',gamma=g,C=c).fit(Xtrainscaled,ytrain)
        print("C:",c)
        print("gamma",g)
        print("Training",KSVM.score(Xtrainscaled,ytrain))
        print("Test",KSVM.score(Xtestscaled,ytest))
```

### From our findings, we see that the best general parameters to use in order to have a good generalizing model running an SVM with an rbf kernel are a C = 10, and a gamma = 0.1

```python
KSVMFinal = SVC(kernel = 'rbf',gamma=0.1,C=10)
KSVMFinal.fit(Xtrainscaled,ytrain)
print("C:",10)
print("gamma",0.1)
print("Training",KSVMFinal.score(Xtrainscaled,ytrain))
print("Test",KSVMFinal.score(Xtestscaled,ytest))
```

### Data reduction using PCA using SVD decomp

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 50)
dataRed = pca.fit_transform(Xdata)
print('# of components:',10)
print(pca.explained_variance_ratio_)
```

```python
var = pca.explained_variance_ratio_
var.sum()
dataRed.shape
```

### By reducing the dataset from 270+ dimensions down to 50, we still preserve 97.6% of the variance and therefore explanatory power of our data for our models. Now we will try and run our models upon this new dataset.

```python
pcaTrain,pcaTest,ytrain,ytest=train_test_split(dataRed,ydatamulti,random_state=0)

model_names = ['logit','rForest','KSVM','LSVM','Tree','KNN']
for i in range(5):
    models[i].fit(pcaTrain,ytrain)
    print('model:',model_names[i])
    print('training score:',models[i].score(pcaTrain,ytrain))
    print('testing score:',models[i].score(pcaTest,ytest))
```

### As expected, we see a decrease in performance across all models using PCA transformed data. This is expected because, ideally, we would never want to use PCA, since we lose variance in our dataset during the reduction in dimensions (not to mention we lose the interpretability of the model). The only reason to use PCA transformation on our data is to reduce the processing load on our computers while losing the minimal amount of explanatory power of our models as possible. However, since we can run the original models just fine, we don't want to be using PCA here.
